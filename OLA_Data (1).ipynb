#### Preprocessing the data

```python
df["Canceled_Rides_by_Customer"] = np.where(df["Incomplete_Rides"]=="No","It's Complete Ride",df["Canceled_Rides_by_Customer"]
```

```python
df["Canceled_Rides_by_Driver"] = np.where(df["Incomplete_Rides"]=="No","It's Complete Ride",df["Canceled_Rides_by_Driver"])
```

```python
df["Incomplete_Rides_Reason"] = np.where(df["Incomplete_Rides"]=="No","It's Complete Ride",df["Incomplete_Rides_Reason"])
```

```python
df["Incomplete_Rides"] = np.where(df["Incomplete_Rides"].isna(),"Ride didn't start",df["Incomplete_Rides"])
```

```python
df["Incomplete_Rides_Reason"] = np.where(df["Incomplete_Rides_Reason"].isna(),"Ride didn't start",df["Incomplete_Rides_Reason"])
```

```python
df["V_TAT"] = np.where(df["Incomplete_Rides_Reason"]=="Ride didn't start",0,df["V_TAT"])
```

```python
df["C_TAT"] = np.where(df["Incomplete_Rides_Reason"]=="Ride didn't start",0,df["C_TAT"])
```

```python
df["Driver_Ratings"] =np.where(df["Incomplete_Rides_Reason"]=="Ride didn't start",0,df["Driver_Ratings"])
```

```python
df["Customer_Rating"] =np.where(df["Incomplete_Rides_Reason"]=="Ride didn't start",0,df["Customer_Rating"])
```

```python
df["Payment_Method"] = np.where(df["Incomplete_Rides_Reason"]=="Ride didn't start","Not Applicable",df["Payment_Method"])
```

```python
df["Canceled_Rides_by_Customer"] = df["Canceled_Rides_by_Customer"].fillna("Unknown")
```

```python
df["Canceled_Rides_by_Driver"] = df["Canceled_Rides_by_Driver"].fillna("Unknown")
```

```python
df.drop(columns=["Booking_ID","Customer_ID","Vehicle Images","Unnamed: 20"],inplace=True)
```

#### Handling the date column

```python
df["Date"] = pd.to_datetime(df["Date"],errors="coerce")
```

#### Make new Category

```python
df["hours"] = df["Date"].dt.hour
```

```python
time_slots={
    0:"Late Night",
    1:"Midnight",
    2:"Midnight",
    3:"Early Morning",
    4:"Early Morning",
    5:"Early Morning",
    6:"Morning",
    7:"Morning",
    8:"Morning",
    9:"Morning",
    10:"Morning",
    11:"Morning",
    12:"After Noon",
    13:"After Noon",
    14:"After Noon",
    15:"After Noon",
    16:"Evening",
    17:"Evening",
    18:"Evening",
    19:"Night",
    20:"Night",
    21:"Night",
    22:"Late Night",
    23:"Late Night"
}
```

```python
df["Time_Slot"] = df["hours"].map(time_slots)
```

#### Make new category

```python
df["Week_Name"] = df["Date"].dt.day_name()
```

```python
df["Weekend/Weekdays"] = np.where(df["Week_Name"].isin(["Saturday","Sunday"]),"Weekend","Weekday")
```
